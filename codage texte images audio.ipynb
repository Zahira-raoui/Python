```python
# convertir une texte en binaire:
def conv_text_bin(text):
    chaine=""
    for i in text:
        c=format(ord(i),'08b')
        chaine+="".join(c)
    return chaine
```

```python
# convertir un fichier texte en binaire:
def fichier_binaire():
    file=open("texte.txt","r")
    for i in file:
        k=conv_text_bin(i)
        return k
```

```python
def decodage(chaine_binary):
    text=""
    for i in range(0,len(chaine_binary),8):
        chaine_8=chaine_binary[i:i+8]#prendre chaque 8 bits
        binary_int = int(chaine_8, 2)#convertir en decimal
        text+=chr(binary_int)#convertir en caractère ASCII 
    return text
```

```python
def list_array(A):
    L=[]
    [nl,nc]=np.shape(A)
    for i in range(nl):
        for j in range(nc):
            L.append(A[i][j])
    return L
```

```python
# codage d'image RGB
def RGB_toBinaire(image):
    I=plt.imread(image)
    r,g,b=I[:,:,0],I[:,:,1],I[:,:,2]# 3 plan 
    r1=list_array(r)
    g1=list_array(g)
    b1=list_array(b)
    r2=codage_image(r1)
    g2=codage_image(g1)
    b2=codage_image(b1)
    return r2,g2,b2
```

```python
RGB_toBinaire("image.jpg")
```

```python
# decodage image RGB a partir des 3 list binaire
```

```python
def decodage_RGB(r1,g1,b1):
    r2=decodage_image_gris(r1)
    g2=decodage_image_gris(g1)
    b2=decodage_image_gris(b1)
    r3=decodage_bianry(r2)
    g3=decodage_bianry(g2)
    b3=decodage_bianry(b2)
    r4=decodage_bianry(r3)
    g4=decodage_bianry(g3)
    b4=decodage_bianry(b3)
    return r4,g4,b4
```

```python
r,g,b=RGB_toBinaire("image.jpg")
decodage_RGB(r,g,b)
```

```python
#fusionner les 3 plans et afficher l'image
r1,g1,b1=decodage_RGB(r,g,b)
k=np.dstack((r1,b1,g1))
plt.imshow(k)
```

```python
#codage audio
def lect_audio(fichier):
    fich=wave.open(fichier,"rb")
    x=fich.readframes(-1)#en hexadecimal
    f=np.frombuffer(x,np.int16).astype('float16')
    min1=np.min(f)
    max1=np.max(f)
    extrait_new=f+(-min1)
    extrait_new1=extrait_new/np.max(extrait_new)
    extrait_new2=(extrait_new1*(pow(2,15))).astype(int)
    return extrait_new2,np.max(extrait_new),min1
```

```python
# converir en binaire
def codage(listo):
    l=[]
    for i in listo:
        k=(format(i,'016b'))
        l.append(k)
    return l
```

```python
listo,n,m=lect_audio("america.wav")
codage(listo)
```

```python
#convertir de binaire vers decimal
def convert_bin_todecimal(listo):
    l=[]
    for i in listo:
        k=int(i,2)
        l.append(k)
    return l
```

```python
lista=codage(listo)
convert_bin_todecimal(lista)
```

```python
#decodage
def decodage(listo,max1,min1):
    n=np.array(listo)/pow(2,15)
    n=(n*max1)+min1
    return n
```

```python
k1=convert_bin_todecimal(lista)
l1,l2,l3=lect_audio("america.wav")
decodage(k1,l2,l3)
```

```python
n=decodage(k1,l2,l3)
sd.play(n)# ecouter audio de signel decode 
```

```python
def NRZI(signal):
    """Encode un signal binaire selon le code NRZI"""
    signal_cop=signal.copy()
    encoded_signal = []
    if signal_cop[0] == 0:
        encoded_signal += V1
    else:
        encoded_signal += V2
    current_state = encoded_signal
    signal_cop.pop(0)
    for bit in signal_cop:
        if bit == 1:
            encoded_signal += current_state
        else:
            current_state = [-x for x in current_state]
            encoded_signal += current_state
    return encoded_signal
```

```python
def decodage_RZI(signal):
    bits = []
    bit_duration = len(V1)# Durée d'un bit dans le signal encodé
    bit_signal = signal[0:0+bit_duration]
    if sum(bit_signal) > 0:
            bits.append(1)
    else:
            bits.append(0)
    for i in range(bit_duration, len(signal), bit_duration):
        bit_signal = signal[i:i+bit_duration]
        if signal[i-1]==signal[i]:
            bits.append(1)
        else:
            bits.append(0)
    return bits
```

```python
def manchester(signal):
    """Encode un signal binaire selon le code Manchester"""
    encoded_signal = []
    for bit in signal:
        if bit == 0:
            encoded_signal += V1
            encoded_signal += V2
        else:
            encoded_signal += V2
            encoded_signal += V1
    return encoded_signal
```

```python
def AMI(signal):
    c=[]
    u=0
    for i in range(0,len(signal)):
        if signal[i]==0:
            c+=V3
        if signal[i]==1:
            if u%2==0:
                c+=V2
            if u%2!=0:
                c+=V1
            u+=1
    return c
```

```python
def Manchester_diff(s):
    c=[]
    if s[0]==0:
        c+=V1
        c+=V2
    else:
        c+=V2
        c+=V1
    j=len(V1)+len(V2)
    for i in range(1,len(s)):
        if s[i]==0:
            c+=c[j-(len(V1)+len(V2)):j]
        if s[i]==1:
            k=c[j-(len(V1)+len(V2)):j]
            for i in range(len(k)):
                k[i]=-k[i]
            c+=k
        j+=len(V1)+len(V2)
    return c
```

```python
def decode_manchester_diff(signal):
    bits = []
    bit_duration = len(V1)+len(V2)   # Durée d'un bit dans le signal encodé
    if signal[0]<0:
        bits.append(0)
    else:
         bits.append(1)
    for i in range(bit_duration, len(signal), bit_duration):
        if signal[i-bit_duration:i]==signal[i:i+bit_duration]:
            bits.append(0)
        else:
            bits.append(1)
    return bits
```

```python
def decode_miller(c):
    bits = []
    bit_duration = len(V1)+len(V2)
    bit_duration1=len(V1)
    for i in range(0,len(c)-1,bit_duration):
        if c[i]==c[i+bit_duration1]:
            bits.append(0)
        elif c[i]!=c[i+bit_duration1]:
            bits.append(1)
    return bits
```

```python
def CMI(s):
    c=[]
    j=0
    if s[0]==0:
        c+=V1
        c+=V1
    else:
        c+=V1
        c+=V2
    j+=len(V1)+len(V2)
    for i in range(1,len(s)):
        if s[i]==1:
            c+=V1
            c+=V2
            j+=2
            
        elif s[i]==0:
            if s[i-1]!=0:
                c+=c[j-(len(V1)):j]
                c+=c[j-(len(V1)):j]
                
            elif s[i-1]==0:
                
                k1=c[j-(len(V1)+len(V2)):j]
                for i in range(len(k1)):
                    k1[i]=-k1[i]
                c+=k1
                j+=len(V1)+len(V2)
    return c
```
